```python
import requests
from bs4 import BeautifulSoup as bs4
```

```python

arxiv_api = 'http://export.arxiv.org/api/query/'
my_params = {
    'search_query':'math:Dwyer, Sagan',
#     'author':'Dwyer'
    'sortBy':'lastUpdatedDate',
    'sortOrder':'descending'
            }
```

```python
data = requests.get(arxiv_api, params=my_params)
```

```python
data
```

```python
arxiv = bs4(data.content, 'lxml')
```
